# Convert drug category values from str to int

````python
drug_list = (['alcohol','amphet','amyl','benzos','caff','cannabis','choc','coke','crack',
              'ecstasy','heroin','ketamine','legalh','lsd','meth','mushrooms','nicotine','semer','vsa'])
````

````python
def parse_drug_value(CL_val):
    CL_vals = CL_val.split("CL")
    #print(CL_vals)
    return int(CL_val[-1])
````

````python
for column in df[drug_list]:
    df[column] = df[column].apply(parse_drug_value).astype('int16')
````
